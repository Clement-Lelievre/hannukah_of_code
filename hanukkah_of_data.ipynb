# Challenge 2

```python
df = pd.read_csv('data/noahs-customers.csv')
df.head()
```

I'm looking for the phone # of a customer whose initials are JD, who bought at least once in 2017 (possibly coffee and/or bagel)

## filter on initials

```python
def get_initials(name:str) -> str:
    name = name.lower().strip().split()
    return ''.join(elem[0] for elem in name)
    

df['initials'] =df.name.apply(get_initials)
df = df[df.initials == 'jd']
df.head()
```

## join tables to get purchase history

```python
products = pd.read_csv('data/noahs-products.csv')
products.head()
```

```python
order_items = pd.read_csv('data/noahs-orders_items.csv')
order_items.head()
```

```python
orders = pd.read_csv('data/noahs-orders.csv')
orders.head()
```

```python
orders_full = orders.merge(order_items[['orderid', 'sku']], on='orderid')
orders_full = orders_full.merge(products[['sku', 'desc']], on='sku')
orders_full = orders_full[['customerid', 'ordered','desc']]
orders_full.head()
```

```python
orders_full['year'] = orders_full.ordered.apply(lambda x : str(x)[:4])
orders_full.head()
```

```python
orders_full = orders_full.merge(df[['customerid', 'initials', 'phone']], on='customerid')
orders_full.head()
```

```python
orders_full
```

```python
orders_full.desc = orders_full.desc.apply(str.lower)
orders_full_filtered = orders_full[(orders_full.year == '2017') & (orders_full.desc.str.contains('coffee') | orders_full.desc.str.contains('bagel'))]
orders_full_filtered
```

answer is `212-771-8924` because the other item descritions are accessories, not coffee or bagel per se

# Challenge 4

we're looking for a woman, client of noah's, who comes very early in the morning, and shopped at Noah's in 2017, and we assume she lives in NY

```python
order_items = pd.read_csv('data/noahs-orders_items.csv')
orders = pd.read_csv('data/noahs-orders.csv')
customers = pd.read_csv('data/noahs-customers.csv')
products = pd.read_csv('data/noahs-products.csv')
```

```python
orders.head(1)
```

```python
order_items.head(1)
```

```python
products.head(1)
```

```python
df = orders.merge(order_items[['orderid', 'sku']], on ='orderid')
df = df.merge(customers, on = 'customerid')
df = df.merge(products,on = 'sku')
df.head()
```

```python
df.ordered =pd.to_datetime(df.ordered)
df['order_year'] = df.ordered.apply(lambda x: x.year)
df['order_hour']=df.ordered.apply(lambda x: x.hour)
df['order_month']=df.ordered.apply(lambda x: x.month)
df = df[(df.order_year == 2018) & (df.order_hour.between(3,4)) & (df.order_month.between(1,6)) & (df.total < 50) & (df.ordered==df.shipped)]
df.shape
```
